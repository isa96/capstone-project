## **Total Covid Test**

```python
# code here
df_tct = df.groupby(['createdAt', 'province', 'covidTestStatus'])['covidTestStatus'].agg(['count'])
df_tct
```

```python
# code here
df_tct = df.groupby(['createdAt', 'province', 'covidTestStatus'])['covidTestStatus'].agg(['count']).reset_index()
df_tct
```

```python
df_tct[df_tct['createdAt'] == "2021-08-02"]
```

```python
# ['createdAt', 'province', 'covidTestStatus']
df['totalNotTestedProvince'] = np.nan
df['totalAlreadyTestedProvince'] = np.nan

dtct_n =  df_tct[df_tct['covidTestStatus'] == "Belum"].copy()
dtct_p = df_tct[df_tct['covidTestStatus'] == "Sudah"].copy()

for i in range(len(dtct_n['createdAt'])):
  df.loc[(df['createdAt'] == dtct_n.iloc[i, 0]) & (df['province'] == dtct_n.iloc[i, 1]), ['totalNotTestedProvince']] = dtct_n.iloc[i, 3]

for i in range(len(dtct_p['createdAt'])):
  df.loc[(df['createdAt'] == dtct_p.iloc[i, 0]) & (df['province'] == dtct_p.iloc[i, 1]), ['totalAlreadyTestedProvince']] = dtct_p.iloc[i, 3]
```
