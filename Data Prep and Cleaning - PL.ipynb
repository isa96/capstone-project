## **Make vaksinStatus from userStatus**

```python
df.head()
```

```python
df.columns
```

userStatus = status warna seseorang saat mengunjungi suatu tempat (hijau/merah/hitam/kuning)

	- hijau: sudah vaksin 2x
	- merah: belum vaksin
	- hitam: positif 
	- kuning: baru vaksin 1x

hijau, kuning --> sudah vaksin

hitam, merah --> belum vaksin

```python
def vaksinTestStatus(s):
    if (s['userStatus'] == "green") or (s['userStatus'] == "yellow"):
        return "Sudah Vaksin"
    else:
        return "Belum Vaksin"
```

```python
df['vaksinTestStatus'] = df.apply(vaksinTestStatus, axis=1)
```

## **Total Vaksin**

```python
# code here
df_tv = df.groupby(['createdAt', 'Provinsi', 'vaksinTestStatus'])['vaksinTestStatus'].agg(['count'])
df_tv
```

```python
# code here
df_tv = df.groupby(['createdAt', 'Provinsi', 'vaksinTestStatus'])['vaksinTestStatus'].agg(['count']).reset_index()
df_tv
```

```python
# ['createdAt', 'Provinsi', 'vaksinTestStatus']
df['totalAlreadyVaccinatedProvince'] = np.nan
df['totalNotVaccinatedProvince'] = np.nan

dtv_n =  df_tv[df_tv['vaksinTestStatus'] == "Belum Vaksin"].copy()
dtv_p = df_tv[df_tv['vaksinTestStatus'] == "Sudah Vaksin"].copy()

for i in range(len(dtv_n['createdAt'])):
  df.loc[(df['createdAt'] == dtv_n.iloc[i, 0]) & (df['Provinsi'] == dtv_n.iloc[i, 1]), ['totalAlreadyVaccinatedProvince']] = dtv_n.iloc[i, 3]

for i in range(len(dtv_p['createdAt'])):
  df.loc[(df['createdAt'] == dtv_p.iloc[i, 0]) & (df['Provinsi'] == dtv_p.iloc[i, 1]), ['totalNotVaccinatedProvince']] = dtv_p.iloc[i, 3]
```
